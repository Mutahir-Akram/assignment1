## Question 1 — Arithmetic & Assignment Operators
In this question I will create two variables and use Python operators on them.

```python
# I chose num1 = 25 and num2 = 6 for demonstration
num1 = 25
num2 = 6

print("==== Arithmetic Operations ====")
print(f"num1 = {num1}, num2 = {num2}")
print("-" * 35)
print(f"Addition       : {num1} + {num2} = {num1 + num2}")
print(f"Subtraction    : {num1} - {num2} = {num1 - num2}")
print(f"Multiplication : {num1} * {num2} = {num1 * num2}")
print(f"Division       : {num1} / {num2} = {num1 / num2:.4f}")
print(f"Floor Division : {num1} // {num2} = {num1 // num2}")
print(f"Modulus        : {num1} % {num2} = {num1 % num2}")
print(f"Exponentiation : {num1} ** {num2} = {num1 ** num2}")
```

```python
# demonstrating compound assignment operators on num1
num1 = 25
print("Starting value of num1:", num1)

# += operator
num1 += 10
print("num1 after += 10 :", num1)   # 35

# -= operator
num1 -= 7
print("num1 after -= 7  :", num1)   # 28

# *= operator
num1 *= 3
print("num1 after *= 3  :", num1)   # 84

# /= operator
num1 /= 6
print("num1 after /= 6  :", num1)   # 14.0
```

## Question 2 — Comparison & Logical Operators
Taking two integers from the user and applying comparison and logical operators.

```python
# user input
val1 = int(input("Please enter first integer: "))
val2 = int(input("Please enter second integer: "))

print(f"\nYou entered: val1 = {val1}, val2 = {val2}")

# all six comparison operators
print("\n--- Comparison Operators ---")
print(f"val1 == val2  -->  {val1 == val2}")
print(f"val1 != val2  -->  {val1 != val2}")
print(f"val1 >  val2  -->  {val1 > val2}")
print(f"val1 <  val2  -->  {val1 < val2}")
print(f"val1 >= val2  -->  {val1 >= val2}")
print(f"val1 <= val2  -->  {val1 <= val2}")
```

```python
# logical operators combining conditions
print("\n--- Logical Operators ---")

# both positive
result_and = val1 > 0 and val2 > 0
print(f"(val1 > 0) and (val2 > 0)  -->  {result_and}")

# at least one is even
result_or = (val1 % 2 == 0) or (val2 % 2 == 0)
print(f"(val1 is even) or (val2 is even)  -->  {result_or}")

# val1 is NOT equal to val2
result_not = not(val1 == val2)
print(f"not(val1 == val2)  -->  {result_not}")
```

## Question 3 — Bitwise Operators
Bitwise operators work on binary representations of integers.

```python
# choosing two integers
m = 9    # 1001 in binary
n = 6    # 0110 in binary

print(f"m = {m}  (binary: {bin(m)})")
print(f"n = {n}  (binary: {bin(n)})")
print()

# AND - both bits must be 1
print(f"m AND n   (m & n)   = {m & n}   binary: {bin(m & n)}")

# OR - at least one bit must be 1
print(f"m OR n    (m | n)   = {m | n}  binary: {bin(m | n)}")

# XOR - bits must be different
print(f"m XOR n   (m ^ n)   = {m ^ n}  binary: {bin(m ^ n)}")

# NOT - flips all bits
print(f"NOT m     (~m)      = {~m}")

# left shift by 2 positions
print(f"m << 2             = {m << 2}  binary: {bin(m << 2)}")

# right shift by 1 position
print(f"m >> 1             = {m >> 1}   binary: {bin(m >> 1)}")
```

## Question 4 — Identity & Membership Operators

```python
# creating a list
my_list = [10, 20, 30, 40, 50, 60]

# identity operators
list_a = my_list          # same object reference
list_b = [10, 20, 30, 40, 50, 60]  # a different object with same data

print("=== Identity Operators ===")
print("list_a is my_list      :", list_a is my_list)       # True
print("list_b is my_list      :", list_b is my_list)       # False
print("list_b is not my_list  :", list_b is not my_list)   # True
print()
print("id(my_list) =", id(my_list))
print("id(list_a)  =", id(list_a),  " <-- same as my_list")
print("id(list_b)  =", id(list_b),  " <-- different object")
```

```python
# membership operators
print("=== Membership Operators ===")
print("List:", my_list)
print()
print("30 in my_list       :", 30 in my_list)       # True
print("75 in my_list       :", 75 in my_list)       # False
print("10 not in my_list   :", 10 not in my_list)   # False
print("99 not in my_list   :", 99 not in my_list)   # True
```

## Question 5 — String Operations

```python
# my chosen sentence
my_string = "Learning Python opens doors to endless opportunities"

print("Original  :", my_string)
print()

# 1. length
print("1. Length of string  :", len(my_string))

# 2. case conversion
print("2. Uppercase         :", my_string.upper())
print("   Lowercase         :", my_string.lower())

# 3. finding a word
word_to_find = "Python"
print(f"3. '{word_to_find}' in string :", word_to_find in my_string)

# 4. replace a word
updated = my_string.replace("Python", "programming")
print("4. After replace     :", updated)

# 5. slicing – first 20 characters
print("5. Slice [0:20]      :", my_string[0:20])

# 6. reverse using slicing
print("6. Reversed          :", my_string[::-1])
```
